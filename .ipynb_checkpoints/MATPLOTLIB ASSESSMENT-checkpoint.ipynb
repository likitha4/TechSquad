```python
import  matplotlib.pyplot as plt
```

```python
x1=[1,3,5,7,9,11,13,15,17] #graph
y1=[2,4,6,8,20,30,25,15,10]
x2=[2,4,6,8,10,12,14,16,18]
y2=[5,10,15,20,25,30,35,40,45]
plt.plot(x1,y1,color='red',label='line1')
plt.plot(x2,y2,color='green',label='line2')
plt.legend()
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')
plt.title('SIMPLE GRAPH')
```

```python
mobiles=['Samsung','iPhone','Redmi Note5','Micromax','Oneplus']
sales=[2000000,15000000,7000000,100000,12500000] #bargraph
plt.bar(mobiles,sales,color='blue')
plt.title('MOBILES VS SALES')
plt.xlabel('Mobile Brands')
plt.ylabel('Sales')
plt.show()
```

```python
slices=[2,3,17,4,1.5] #pie chart
se_eng=['ASK','YAHOO','GOOGLE','BING','BAIDU']
col=['r','b','g','c','y']

plt.pie(slices,labels=se_eng,colors=col,startangle=180,explode=(0,0,0.01,0.1,0),autopct='%1.2f%%')
plt.title('USAGE OF SEARCH ENGINES- WORLD WIDE')
plt.show()
```
